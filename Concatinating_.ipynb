```python
import pandas as pd
import numpy as np
```

```python
pakistan=pd.DataFrame({
  "City":["karachi","lahore","islamabad"],
  "temperature":[50,40,20],
  "humidity":[60,65,70]

})
pakistan
```

```python
india_data=pd.DataFrame({
  "City":["karnataka","ahemdabad","bombay"],
  "temperature":[30,35,40],
  "humidity":[60,65,70]

})

india_data
```

```python

df=pd.concat([pakistan,india_data])
df
```

```python
# to correct index
df_index=pd.concat([pakistan,india_data],ignore_index=True)
df_index
```

```python
df_keys=pd.concat([pakistan,india_data],keys=["pakistan","india"])
df_keys
```

```python
# for column wise concatination usr this method

df_columnwise=pd.concat([pakistan,india_data],axis=1)
df_columnwise
```

```python
#merge option

df_merged= pd.merge(pakistan,india_data, on="City")
df_merged.head()

#the issue here is that it is joining only common names since there are no common names so we use

```

```python
df_merged= pd.merge(pakistan,india_data, on="City", how= "outer")
df_merged.head()
```
